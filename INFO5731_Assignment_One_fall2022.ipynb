	r	V

```python
import pandas as pd
import io
  
df = pd.read_csv(io.BytesIO(uploaded['Assignment1_denton_housing.csv']))
df.head()
```

```python
rslt_df = df[df['title_field'] == "Occupied Housing Units"]
res = rslt_df.drop(["year","title_field"],axis=1)
```

```python
final = 0-res.diff()
final = final.dropna()
final=final.iloc[::-1]
```

```python
final["year"] = ['2008-2009','2009-2010','2010-2011','2011-2012','2012-2013','2013-2014']
print("difference in occupied units")
final
```

```python
output=final.sort_values(['value'],ascending=True)
print("difference in occupied units in ascending order")
output
```

```python
totval=df.loc[df['title_field'] == 'Total Housing Units']
totval.reset_index(drop=True,inplace=True)
vacval=df.loc[df['title_field'] == 'Vacant Housing Units']
vacant = vacval.drop(["year","title_field",],axis =1)
vacant.reset_index(drop=True,inplace=True)
totval.insert(3, "value_of_vacant_units", vacant)
occval=df.loc[df['title_field'] == 'Occupied Housing Units']
occupied = occval.drop(["year","title_field"],axis =1)
occupied.reset_index(drop=True,inplace=True)
totval.insert(4, "value_of_occupied_units", occupied)
vacant_percentage=(totval['value_of_vacant_units']/totval['value'])*100
occupied_percentage=(totval['value_of_occupied_units']/totval['value'])*100
totval.insert(5, "perventage_of_vacant_units", vacant_percentage)
totval.insert(6, "perventage_of_occupied_units", occupied_percentage)
totval
```
